```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def histogram_specification(img, target_hist):
    hist, bins = np.histogram(img.ravel(), 256, [0,256])
    cdf = hist.cumsum()
    cdf = 255 * cdf / cdf[-1]
    new_img = np.interp(img.ravel(), bins[:-1], cdf).reshape(img.shape)
    return new_img
```

```python
# Contoh target histogram
target_hist = np.zeros(256)
target_hist[0:50] = np.linspace(0,255,50)
target_hist[50:100] = 255
target_hist[100:] = np.linspace(255,0,156)
```

```python
# Baca gambar dengan mode grayscale
img = cv2.imread('images/paramore2.jpg', 0)
```

```python
# Aplikasikan histogram specification atau histogram equalization
new_img = histogram_specification(img, target_hist) # atau histogram_equalization(img)
```

```python
# Tampilkan histogram asli dan histogram baru
plt.subplot(2,1,1)
plt.hist(img.ravel(), 256, [0,256])
plt.subplot(2,1,2)
plt.hist(new_img.ravel(), 256, [0,256])
plt.show()
```

```python
# Tampilkan gambar asli dan gambar baru
plt.subplot(2,1,1)
plt.imshow(img, cmap='gray')
plt.subplot(2,1,2)
plt.imshow(new_img, cmap='gray')
plt.show()
```
